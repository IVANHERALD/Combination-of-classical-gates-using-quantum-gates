```python
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
```

```python
from qiskit.visualization import *
```

MULTIPLEXER

```python
qc=QuantumCircuit(13,1)
```

```python
qc.mct([0,1,5],6)

qc.x(1)
qc.mct([0,1,3],7)

qc.x(0)
qc.mct([0,1,2],8)

qc.x(1)
qc.mct([0,1,4],9)

qc.barrier()
qc.ccx(6,7,10)
qc.cx(6,7)
qc.cx(7,10)

qc.ccx(8,10,11)
qc.cx(10,8)
qc.cx(8,11)

qc.ccx(9,11,12)
qc.cx(11,9)
qc.cx(9,12)
qc.measure(12,0)
qc.barrier()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1024, memory=True)
output = job.result().get_memory()[0]
```

```python
print(output)
```

D Flip Flop

```python
qc=QuantumCircuit(7,1)

qc.x(0)      #clock
# qc.x(1)    #D
qc.x(4)      #Q

qc.ccx(0,1,2)
qc.x(2)
qc.x(1)
qc.ccx(0,1,3)
qc.x(3)
qc.x(1)
qc.ccx(3,4,5)
qc.x(5)
qc.ccx(2,5,4)
qc.x(4)
qc.measure(4,0)
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1024, memory=True)
output = job.result().get_memory()[0]
```

```python
print(output)
```

COMPLETE CIRCUIT

```python
qc=QuantumCircuit(18,1)


# s0 - q0
# s1 - q1
# I0 - q2
# I1 - q3
# I2 - q4
# I3 - q5

# S0 , S1 , I3 - q6
# S1' , S0 , I1 - q7
# S1' , S0' , I0 - q8
# S1 , S0' , I2 - q9


# multiplexer output - q12
# D flip flop D      - q13
# Q                  - q16


# s0 - 0
# s1 - 0
# I0 - 1
# y  - 1  --->multiplexer output

qc.x(2)
qc.x(13)

qc.mct([0,1,5],6)

qc.x(1)
qc.mct([0,1,3],7)

qc.x(0)
qc.mct([0,1,2],8)

qc.x(1)
qc.mct([0,1,4],9)

qc.barrier()
qc.ccx(6,7,10)
qc.cx(6,7)
qc.cx(7,10)

qc.ccx(8,10,11)
qc.cx(10,8)
qc.cx(8,11)

qc.ccx(9,11,12)
qc.cx(11,9)
qc.cx(9,12)

qc.barrier()



qc.ccx(12,13,14)
qc.x(14)
qc.x(13)
qc.ccx(12,13,15)
qc.x(13)
qc.x(15)
qc.ccx(15,16,17)
qc.x(17)
qc.ccx(14,17,16)
qc.x(16)
qc.measure(16,0)

qc.draw('mpl')

```

```python
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1024, memory=True)
output = job.result().get_memory()[0]
```

```python
print(output)
```
